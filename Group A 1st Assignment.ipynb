```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("darkgrid")
```

```python
def linear(x):
    return x

x = np.linspace(-10,10,1000)
plt.axvline(x=0, c="black")
plt.axhline(y=0, c='black')
plt.plot(x,linear(x), label = "Line 1")
plt.plot(x,linear(x)*2, label = "Line 2")
plt.plot(x,linear(x)*3, label = "Line 3")
plt.title("Identity Linear Activation Function")
plt.show()
```

```python
def pclinear(x):
    x1=[]
    for i in x:
        A=-1
        B=1
        if(i<=A):
            x1.append(-1)
        elif(A<i<B):
            x1.append(i)
        else:
            x1.append(1)
    return x1

x = np.linspace(-10,10,1000)
plt.plot(x,pclinear(x))
plt.axvline(x=0, c="black")
plt.axhline(y=0, c='black')
plt.title("Piecewise Linear Activation Function")
plt.show()
```

```python
def threshold(x):
    threshold=0
    return (x >= threshold).astype(int)

x = np.linspace(-10,10,1000)
plt.plot(x,threshold(x))
plt.xlim(-10,10)
plt.title("Unipolar Threshold Activation Function")
plt.show()
```

```python
def bipolar_threshold(x):
    return np.where(x >= 0 , +1,-1)

x = np.linspace(-10,10,1000)
plt.plot(x,bipolar_threshold(x))
plt.axhline(y=0, c= "black")
# plt.axvline(x=0, c= "black")
plt.title("Bipolar Threshold Activation Function")
plt.show()
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.linspace(-10,10,1000)
y = np.linspace(-20,20,1000)
z = np.linspace(-5,5,1000)
plt.plot(x,sigmoid(x))
plt.plot(x,sigmoid(y))
plt.plot(x,sigmoid(z))
plt.axhline(y=0, c= "black")
plt.axvline(x=0, c= "black")
plt.title("Sigmoid Non-Linear Activation Function")
plt.show()
```

```python
def hyperbolic_tangent(x):
    return (1 - np.exp(-x)) / (1 + np.exp(-x))

x = np.linspace(-10,10,1000)
y = np.linspace(-20,20,1000)
z = np.linspace(-5,5,1000)
plt.plot(x,hyperbolic_tangent(x))
plt.plot(x,hyperbolic_tangent(y))
plt.plot(x,hyperbolic_tangent(z))
plt.axhline(y=0, c= "black")
plt.axvline(x=0, c= "black")
plt.title("Hyperbolic Tangent Non-Linear Activation Function")
plt.show()
```
